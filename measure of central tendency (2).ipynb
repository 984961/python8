```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import boxcox
import warnings
warnings.filterwarnings('ignore')
```

```python
##generating a random variable
exp=np.random.normal(25000,15000,1000) #(mean, std dev, no.of data points)
exp
```

```python
## plotting the randomly generated exp
#visualizing the distribution
sns.displot(exp, kde=True)
plt.show()
```

```python
##calculating mean
mean_1 = np.mean(exp)
mean_1
```

```python
##calculating median
median_1 = np.median(exp)
print(median_1)
```

```python
##adding one huge point,outlier exp-->1000 datapoint + 1 more dt pt
exp1=np.append(exp,[10000000])
```

```python
sns.displot(exp1)
```

```python
np.mean(exp1)
```

```python
np.median(exp1) #skewed distribution
```

```python
#we can see that after adding the outlier , the mean is shifting towards the outlier
#but median is remaining yhe same
#hence prove that median is the appropreate measure of central tendency
#when we have outlier
```

```python
##scipy is librery
##stats module
from scipy import stats
```

```python
print(stats.skew(exp))
print(stats.skew(exp1))
```

```python
print(stats.kurtosis(exp))
print(stats.kurtosis(exp1))
```

```python
column = "CGPA"
values = data[column]
```

```python
data.columns
```

```python
plt.figure(figsize=(15, 8))
plt.subplot(2, 3, 1)
plt.hist(values, bins=30, color="skyblue", edgecolor="black")
plt.title("Original Distribution")
sns.displot(data, kde=True)
plt.show()
```

```python
#log transformer
log_transformer = np.log1p(values)
plt.subplot(2, 3, 2)
plt.hist(log_transformer, bins=30, color="pink", edgecolor="black")
plt.title("Log Transformed")
sns.displot(data, kde=True)
plt.show()
```

```python
# Box-Cox Transformation (requires positive values)
positive_values = values + 1  # Ensure all values are positive
boxcox_transformed,_= boxcox(positive_values)
plt.subplot(2, 3, 3)
plt.hist(boxcox_transformed, bins=30, color="pink", edgecolor="red")
plt.title("Box-Cox Transformed Distribution")
```

```python
#square root transformation
sqrt_transformed = np.sqrt(values + 1)  # Corrected variable name
plt.subplot(2, 3, 4)
plt.hist(sqrt_transformed, bins=30, color="gold", edgecolor="black")
plt.title("Square Root Transformed Distribution")
```

```python
# reciprocal Transformation
#adding a small constant to avoid division by zero
reciprocal_transformed = 1 / (values +1)
plt.subplot(2, 3, 5)
plt.hist(reciprocal_transformed, bins=30, color="plum", edgecolor="black")
plt.title("Reciprocal Transformed Distribution")
```
